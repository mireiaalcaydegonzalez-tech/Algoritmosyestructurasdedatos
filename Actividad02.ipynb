## Paso 1: Definir la clase Graph

Implementamos dos representaciones: lista de adyacencia y matriz de adyacencia

```python
from collections import deque, defaultdict

class Graph:
    # Grafo no dirigido
    def __init__(self, edges):
        self.edges = edges
        self.adj_list = defaultdict(list)
        self.nodes = set()
        
        # Construir la lista
        for u, v in edges:
            self.adj_list[u].append(v)
            self.adj_list[v].append(u)
            self.nodes.add(u)
            self.nodes.add(v)
        
        # Ordenar para que sea consistente
        for node in self.adj_list:
            self.adj_list[node].sort()
    
    def get_list(self):
        return dict(self.adj_list)
    
    def get_matrix(self):
        nodes_sorted = sorted(self.nodes)
        n = len(nodes_sorted)
        idx = {node: i for i, node in enumerate(nodes_sorted)}
        
        # Inicializar con ceros
        mat = [[0] * n for _ in range(n)]
        
        for u, v in self.edges:
            i, j = idx[u], idx[v]
            mat[i][j] = 1
            mat[j][i] = 1
        
        return nodes_sorted, mat
    
    def dfs(self, start):
        visited = set()
        result = []
        
        def recorrer(node):
            visited.add(node)
            result.append(node)
            for vecino in self.adj_list[node]:
                if vecino not in visited:
                    recorrer(vecino)
        
        recorrer(start)
        return result
    
    def bfs(self, start):
        visited = {start}
        queue = deque([start])
        result = []
        
        while queue:
            node = queue.popleft()
            result.append(node)
            
            for vecino in self.adj_list[node]:
                if vecino not in visited:
                    visited.add(vecino)
                    queue.append(vecino)
        
        return result
    
    def is_connected(self):
        if not self.nodes:
            return True
        
        start = next(iter(self.nodes))
        visited = set()
        
        def explore(n):
            visited.add(n)
            for v in self.adj_list[n]:
                if v not in visited:
                    explore(v)
        
        explore(start)
        return len(visited) == len(self.nodes)
```

## Funciones auxiliares para imprimir los resultados

```python
def print_list(g):
    adj = g.get_list()
    print("\nLista de adyacencia:")
    for node in sorted(adj.keys()):
        print(f"{node}: {adj[node]}")

def print_matrix(g):
    nodes, mat = g.get_matrix()
    print("\nMatriz de adyacencia:")
    print("  ", " ".join(f"{n:2}" for n in nodes))
    for i, node in enumerate(nodes):
        row = " ".join(f"{mat[i][j]:2}" for j in range(len(nodes)))
        print(f"{node} [{row}]")
```

## Apartado (b): Algoritmos de Recorrido

Se implementan DFS y BFS para recorrer el grafo desde un nodo dado,
permitiendo analizar la estructura y conectividad de la red.

## Apartado (c): Verificación de Conectividad

Se analiza si el grafo es conexo, es decir, si todas las estaciones
están conectadas directa o indirectamente.

```python
def print_adjacency_list(graph):
    """Imprime la lista de adyacencia de forma legible."""
    adj_list = graph.get_adjacency_list()
    print("\nLista de adyacencia:")
    for node in sorted(adj_list.keys()):
        neighbors = adj_list[node]
        print(f"{node}: {neighbors}")

def print_adjacency_matrix(graph):
    """Imprime la matriz de adyacencia de forma legible."""
    nodes, matrix = graph.get_adjacency_matrix()
    
    print("\nMatriz de adyacencia:")
    # Encabezado
    print("  ", " ".join(f"{node:2}" for node in nodes))
    
    # Filas
    for i, node in enumerate(nodes):
        row_str = " ".join(f"{matrix[i][j]:2}" for j in range(len(nodes)))
        print(f"{node} [{row_str}]")
```

## Apartado (a): Implementación de Stack usando Queue

Para simular el comportamiento LIFO de una pila,
se inserta el nuevo elemento al final de la cola
y posteriormente se rotan los elementos anteriores.

```python
from collections import deque
class StackUsingQueue:
    """
    Implementación de una pila (Stack) usando una cola (Queue).
    """

    def __init__(self):
        """Inicializa la pila con una cola vacía."""
        self.queue = deque()

    def push(self, item):
        """
        Inserta un elemento en la parte superior de la pila.

        Complejidad: O(n)
        """
        # Añadimos el elemento al final de la cola
        self.queue.append(item)

        # Rotamos los elementos anteriores
        for _ in range(len(self.queue) - 1):
            self.queue.append(self.queue.popleft())

    def pop(self):
        """
        Extrae y devuelve el elemento superior de la pila.

        Complejidad: O(1)
        """
        if self.is_empty():
            raise IndexError("La pila está vacía")

        return self.queue.popleft()

    def peek(self):
        """
        Devuelve el elemento superior sin eliminarlo.

        Complejidad: O(1)
        """
        if self.is_empty():
            raise IndexError("La pila está vacía")

        return self.queue[0]

    def is_empty(self):
        """
        Comprueba si la pila está vacía.

        Complejidad: O(1)
        """
        return len(self.queue) == 0

    def size(self):
        """
        Devuelve el tamaño de la pila.

        Complejidad: O(1)
        """
        return len(self.queue)
```

## Apartado (b): Complejidad temporal

La operación push() tiene complejidad O(n)
porque requiere reorganizar todos los elementos.

La operación pop() tiene complejidad O(1)
ya que extrae directamente el primer elemento.

## Apartado (c): Paréntesis balanceados

Se utiliza la pila implementada anteriormente
para verificar que cada paréntesis abierto
tiene su correspondiente cierre.

```python
def is_balanced_parentheses(expression):
    """
    Comprueba si una expresión matemática tiene
    los paréntesis balanceados.

    Devuelve True si está balanceada y False en caso contrario.

    Complejidad: O(n)
    """
    stack = StackUsingQueue()

    for char in expression:
        if char == "(":
            stack.push(char)

        elif char == ")":
            if stack.is_empty():
                return False

            stack.pop()

    return stack.is_empty()
```

## Casos de prueba

A continuación se prueban distintos casos,

```python
if __name__ == "__main__":
    print("=" * 50)
    print("PRUEBAS STACK")
    print("=" * 50)

    stack = StackUsingQueue()

    stack.push(10)
    stack.push(20)
    stack.push(30)

    print("Top:", stack.peek())       # 30
    print("Pop:", stack.pop())        # 30
    print("Top:", stack.peek())       # 20
    print("Tamaño:", stack.size())    # 2

    print("\n" + "=" * 50)
    print("PRUEBAS PARÉNTESIS")
    print("=" * 50)

    test_cases = [
        ("((2+3)*(4-1))", True),
        ("((2+3)*4-1))", False),
        ("()", True),
        ("(())", True),
        ("((())", False),
        ("())", False),
        ("", True),
    ]

    for expression, expected in test_cases:
        result = is_balanced_parentheses(expression)
        status = "OK" if result == expected else "ERROR"
        print(f"{expression!r} -> resultado={result}, esperado={expected} -> {status}")
        assert result == expected, f"Caso fallido: {expression!r}"
```
